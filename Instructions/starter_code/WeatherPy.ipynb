```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import random
import os
import csv

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = base_url + "appid=" + weather_api_key + "&q="
response = requests.get(query_url + "abcd").json()
city = []
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []
for city in range(len(cities)):
    try:
        print(f"Processing Record: {city} for city: {cities[city]}")
        
        weather_response = requests.get(query_url + cities[city]).json()
        
        city.append(weather_response['name'])
        
        lat.append(weather_response['coord']['lat'])
        
        lng.append(weather_response['coord']['lon'])
        
        max_temp.append(weather_response['main']['temp_max'])
        
        humidity.append(weather_response['main']['humidity'])
        
        cloudiness.append(weather_response['clouds']['all'])
        
        wind_Speed.append(weather_response['wind']['speed'])
        
        country.append(weather_response['sys']['country'])
        
        date.append(weather_response['dt'])
    except Exception:
        print("City not found. Skipping....")
```

```python
# Create Northern and Southern Hemisphere DataFrames

nhem = data[data['Lat'] >= 0] 
nhem.head()
```

```python
shem = data[data['Lat'] <0] 
shem.head()

shem = shem.set_index("City")
shem = shem.drop("paita", axis=0)
shem.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python

x_values = nhem['Lat']
y_values = nhem['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Lat')
plt.ylabel('Max Temp')
plt.show()

print(f"The linear regression shows the farther away from the equator, the colder it is.")
```
